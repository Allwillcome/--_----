```python
import os

def readMotionFile(filename):
    """ Reads OpenSim .sto files.

    Parameters
    ----------

    filename: absolute path to the .sto file

    Returns
    -------

    header: the header of the .sto
    labels: the labels of the columns
    data: an array of the data

    """

    if not os.path.exists(filename):
        print('file do not exists')

    file_id = open(filename, 'r')

    # read header
    next_line = file_id.readline()
    header = [next_line]
    nc = 0
    nr = 0
    while not 'endheader' in next_line:
        if 'datacolumns' in next_line:
            nc = int(next_line[next_line.index(' ') + 1:len(next_line)])
        elif 'datarows' in next_line:
            nr = int(next_line[next_line.index(' ') + 1:len(next_line)])
        elif 'nColumns' in next_line:
            nc = int(next_line[next_line.index('=') + 1:len(next_line)])
        elif 'nRows' in next_line:
            nr = int(next_line[next_line.index('=') + 1:len(next_line)])

        next_line = file_id.readline()
        header.append(next_line)

    # process column labels
    next_line = file_id.readline()
    if next_line.isspace() == True:
        next_line = file_id.readline()

    labels = next_line.split()

    # get data
    data = []
    for i in range(1, nr + 1):
        d = [float(x) for x in file_id.readline().split()]
        data.append(d)

    file_id.close()

    return header, labels, data
```

```python
# Xsens MVN Analysis 四元数转换为欧拉角
import numpy as np
from scipy.spatial.transform import Rotation
Quats = [0.988623246,-0.006829393,-0.012578669,-0.149730472]
Eulers = [-0.558075624,-1.542374693,-17.21688041]
def XsensQuastTranformEulers(Quats):
    rotation = Rotation.from_quat(Quats)
    trans_Eulers = rotation.as_euler("xzy",degrees=True)
    result_Eulers = [trans_Eulers[1],-trans_Eulers[2],-(180+trans_Eulers[0])]

    # 保留 9 位小数
    return np.round(result_Eulers,9).tolist()

transformed_eulers = XsensQuastTranformEulers(Quats)

print("Xsens MVN Analysis  原始欧拉角：{}".format(Eulers))
print("使用scipy经过四元数转换后的欧拉角：{}".format(transformed_eulers))
```

```python
# Xsens MTw 数据 和 Xsens MVN 
# 四元数转换为欧拉角转换结果不一致
Eulers = [-62.631684,-86.936562,-125.544770]
Quats = [-0.034270,-0.695192,0.066517,-0.714919]

transformed_eulers = XsensQuastTranformEulers(Quats)

print("Xsens MVN Analysis  原始欧拉角：{}".format(Eulers))
print("使用scipy经过四元数转换后的欧拉角：{}".format(transformed_eulers))
```

```python
# Xsens MVN Analysis 欧拉角转换为四元数
import numpy as np
from scipy.spatial.transform import Rotation
Quats = [0.988623246,-0.006829393,-0.012578669,-0.149730472]
Eulers = [-0.558075624,-1.542374693,-17.21688041]

def XsnesEulersTransformQuats(Eulers):
    result_Eulers = [Eulers[1],-Eulers[2],-(180+Eulers[0])]
    rotation = Rotation.from_euler("yzx",Eulers,degrees=True)
    trans_Quats = rotation.as_quat()
    result_Quats = [trans_Quats[3],trans_Quats[1],trans_Quats[2],trans_Quats[0]]

    #保留 9 位小数
    return np.round(result_Quats,9).tolist()

transforemed_quats = XsnesEulersTransformQuats(Eulers)

print("Xsens MVN Analysis 原始四元数：{}".format(Quats))
print("使用scipy经欧拉角转换后的四元数： {}".format(transforemed_quats))
```
