```python
# 导入各种安装包
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import jieba
import os
```

```python
# 导入并打开文件夹

image_path = "关于.png"
#文本文件夹位置
path = r"C:\Users\openmind\OneDrive - bsu.edu.cn\SoftwearBackup\Obsidian\人生资本\人生资本_日志.md"
with open(path,"r",encoding="utf8") as f:
    text = f.read()
```

```python
# 自定义结巴分词设置内容
mask = np.array(Image.open(image_path))
font = r"C:Windows\Fonts\simhei.ttf"
sep_list = jieba.lcut(text,cut_all=True)
sep_list = " ".join(sep_list)
stopwords_list = ["不","都要","要","都","的","指数","自己","第","周"]

wc = WordCloud(
    font_path = font,
    margin=2,
    mask=mask,
    background_color="white",
    max_font_size=200,
    max_words=200,
    stopwords=stopwords_list
)
```

```python
#生成并保存词云
wc.generate(sep_list)
wc.to_file("词云.jpg")
```

```python
#显示词云
plt.imshow(wc,interpolation="bilinear")
plt.axis("off")
plt.show()
```
