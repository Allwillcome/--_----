```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#构建函数
x  = np.linspace(1,10,100)
y = np.sin(x)
z = np.cos(x)
t = np.tan(x)
```

```python
#设置画布
# 注意，这里是 axs 复数，下面也是
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(6,9))

#可以简写为
# fig,axs = plt.subplots(3, 1,figsize(6,9))

#也可以不加尺寸
# fig,axs = plt.subplots(3, 1)


#画布和axs 绘图必须同时执行才能显示以下内容

#设置子标题
axs[0].set_title("sin(x)")
axs[0].plot(x,y)

#设置子标题
axs[1].set_title("cos(x)")
axs[1].plot(x,z)

#设置子标题
axs[2].set_title("tan(x)")
axs[2].plot(x,t)

#设置三张图片的总标题
plt.suptitle("sin(x) cos(x) tan(x)")
#如果使用 plt.titile("sin(x) cos(x) tan(x)") 会设置为最后一张图标题
plt.savefig("sin_cos_tan(x)")

plt.show()
```

```python
#设置 1 行 3 列的画布

fig, axs = plt.subplots(nrows=1,ncols=3, figsize=(12,3))


#命名第 1 个子图
axs[0].set_title("sin(x)")
#绘制第 1 个子图
axs[0].plot(x,y)

#命名第 2 个子图
axs[1].set_title("cos(x)")
#绘制第 2 个子图
axs[1].plot(x,z)

#命名第 2 个子图
axs[2].set_title("tan(x)")
axs[2].plot(x,t)

#命名总标题 fontsize 设置字体大小,y 设置距离子标题的纵向距离
plt.suptitle("sin_cos_tan(x)",fontsize=16,y=1.1)
plt.savefig("1 rows 3 cols")
```

```python
# 画布设置为 2 行 2 列
fig, axs = plt.subplots(nrows = 2, ncols=2)

# 多行或者多列就得使用 axs[0][0] 
# 单行或单列则使用 axs[0]
# 不管哪种情况都是使用复数 axs

# 命名左上角子图标题
axs[0][0].set_title("sin(x)")
# 绘制左上角子图
axs[0][0].plot(x,y)

# 命名右上角子图标题
axs[0][1].set_title("cos(x)")
# 绘制右上角子图
axs[0][1].plot(x,z)

# 命名左下角子图标题
axs[1][0].set_title("tan(x)")
axs[1][0].plot(x,t)

#命名右下角子图标题
axs[1][1].set_title("x")
axs[1][1].plot(x,x)

#命名全图名称
plt.suptitle("sin_cos_tan_x")
```
