```python
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d

def DataFrameInterp1d(df,joints):
    # 将数据框转换为 NumPy 数组
    arr = df.to_numpy()

    # 设置新数组长度为 101
    new_length = 101

    # 生成新的 X 坐标轴
    x_old = np.arange(len(arr))
    x_new = np.linspace(0, len(arr) - 1, new_length)

    # 对每一列数据进行线性插值
    new_arrs = []
    for i in range(arr.shape[1]):
        f = interp1d(x_old, arr[:, i], kind='linear')
        new_arr = f(x_new)
        new_arrs.append(new_arr)

    # 将新数组转换为数据框
    norm_df = pd.DataFrame(np.column_stack(new_arrs), columns=joints)
    return norm_df

```

```python
Right_norm_df = DataFrameInterp1d(Right_df,Right_joints)
Left_norm_df = DataFrameInterp1d(Left_df,Left_joints)
```

## 髋关节图片绘制

```python
side  = "Right"
joint = "Hip"
Right_Flexion = side + " "+ joint + " "+ "Flexion/Extension"
Right_Abduction = side+ " "+ joint + " "+ "Abduction/Adduction"
Right_Internal = side+ " "+ joint + " "+ "Internal/External Rotation"

side = "Left"
Left_Flexion = side + " "+ joint + " "+ "Flexion/Extension"
Left_Abduction = side+ " "+ joint + " "+ "Abduction/Adduction"
Left_Internal = side+ " "+ joint + " "+ "Internal/External Rotation"


fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12,3),dpi=500)

Left_norm_df[Left_Flexion].plot(ax=axs[0],color="pink",linestyle = "--",label="患侧")
Left_norm_df[Left_Abduction].plot(ax=axs[1],color="pink",linestyle = "--",label="患侧")
Left_norm_df[Left_Internal].plot(ax=axs[2],color="pink",linestyle = "--",label="患侧")

Right_norm_df[Right_Flexion].plot(ax=axs[0],label="健侧")
Right_norm_df[Right_Abduction].plot(ax=axs[1],label="健侧")
Right_norm_df[Right_Internal].plot(ax=axs[2],label="健侧")



plt.rcParams['font.sans-serif'] = ['Heiti TC']
plt.rcParams['font.family'] = ['Heiti TC']
plt.rcParams['axes.unicode_minus'] = False

for ax in axs:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_linewidth(1)
    ax.spines['left'].set_linewidth(False)
    ax.set_xlim(0,100)
    ax.tick_params(axis='y', length=0, width=0)
    ax.grid(axis='y', linestyle='--', linewidth=0.5)
    ax.legend()

axs[1].set_xlabel("步态周期：同侧脚从足跟首次触地到下次触地")
axs[0].set_ylabel("角度(°)")

axs[0].set_title("屈曲/伸展角度")
axs[1].set_title("内收/外展角度")
axs[2].set_title("内旋/外旋角度")

plt.suptitle("髋关节角度",y=1.1,fontsize=16)
```
