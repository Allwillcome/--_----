```python
joints = ["颅顶点","第七颈椎","胸骨下点","脐点","耻骨","左肩峰","左肘","左腕","左中指尖","左髋","左膝","左踝","左足跟","左趾尖","右肩峰","右肘",
 "右腕","右中指尖",	"右髋",	"右膝",	"右踝",	"右足跟","右趾尖","重心"]	
three_D = ["x","y","z"]
joint_3D = [joint + " " + one_D for joint in joints for one_D in three_D ]
joint_3D
columns = joint_3D + [""]
```

```python
def bio_tsv_to_csv(filename):
    with open(filename,"rb") as f:
        result = chardet.detect(f.read())
    encoding = result["encoding"]
    df = pd.read_csv(filename,encoding=encoding,skiprows=10,sep="\t",header=None)
    df.columns = columns
    name, ext = os.path.splitext(filename)
    df.to_csv(name+".csv",index=None)
    return df
```

```python
bio_tsv_to_csv(filename)
```
