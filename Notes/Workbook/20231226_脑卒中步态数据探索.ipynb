```python
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# 找到峰值
peaks, _= find_peaks(df["Left Foot x"], height=25, distance=200)

# 绘制原始数据
plt.figure(figsize=(6,3))
plt.plot(df["Left Foot x"], label="Left Foot x")
plt.title('Left Foot x with peaks')

# 在峰值位置用圆圈标记
plt.plot(peaks, df["Left Foot x"].iloc[peaks], "o", label="peaks")
plt.legend(loc='best')
plt.show()
```
