```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#æž„å»ºå‡½æ•°
x  = np.linspace(1,10,100)
y = np.sin(x)
z = np.cos(x)
t = np.tan(x)
```
