```python
def plot_figure_L(path, file, a, b,zy):
    #Á°ÆÂÆöÂõæÁâáÂêçÁß∞Âíå‰øùÂ≠òÂêçÁß∞
    regex = r"BioLab_(.*)_(.*)"
    mysearch = re.search(regex,file)
    if mysearch:
        plot_name = mysearch.group(1)
    path_plot = os.path.join(path,"plot",plot_name+"_"+zy)
    title_plot = plot_name+"_"+ zy
    path_plot_sum = os.path.join(path,"plot",plot_name + "_"+ zy+"_Sum")
    tile_plot_sum = plot_name + "_"+ zy + "_Sum"

    #ËØªÂèñÊñá‰ª∂
    df = pd.read_csv(file)
    df_filter = df[df["part"]==zy]
    p1 = df_filter["p1-c"].loc[a:b]
    p2 = df_filter["p2-c"].loc[a:b]
    p3 = df_filter["p3-c"].loc[a:b]
    p4 = df_filter["p4-c"].loc[a:b]
    p5 = df_filter["p5-c"].loc[a:b]
    p6 = df_filter["p6-c"].loc[a:b]
    p7 = df_filter["p7-c"].loc[a:b]
    p8 = df_filter["p8-c"].loc[a:b]
    p9 = df_filter["p9-c"].loc[a:b]
    p10 = df_filter["p10-c"].loc[a:b]
    p11 = df_filter["p11-c"].loc[a:b]
    p12 = df_filter["p12-c"].loc[a:b]

    fig1 = plt.figure(1)
    # plt.ticklabel_format(style="plain")
    plt.plot(p1,label="p1",color="lime")
    plt.plot(p2,label="p2")
    plt.plot(p3,label="p3")
    plt.plot(p4,label="p4")
    plt.plot(p5,label="p5")
    plt.plot(p6,label="p6")
    plt.plot(p7,label="p7")
    plt.plot(p8,label="p8")
    plt.plot(p9,label="p9")
    plt.plot(p10,label="p10")
    plt.plot(p11,label="p11")
    plt.plot(p12,label="p12",color="black")

    plt.legend(loc=2, bbox_to_anchor=(1.05,1.0),borderaxespad = 0.)
    #Â∞ÜÂõæ‰æãÊîæÁΩÆÂú®ÂõæË°®Â§ñ
    plt.title(title_plot)
    plt.grid(linestyle="-.")
    # plt.tight_layout()
    plt.savefig(path_plot, dpi=600,bbox_inches='tight')

    fig2 = plt.figure(2)
    # plt.ticklabel_format(style="plain")
    p_sum = p1+p2+p3+p4+p5+p6+p7+p8+p9+p10+p11+p12
    plt.plot(p_sum);
    plt.title(tile_plot_sum)
    plt.grid(linestyle="-.")
    plt.savefig(path_plot_sum, dpi=600,bbox_inches='tight')
    # plt.tight_layout()
    plt.show()
    plt.close()

def plot_figure_R(path, file, a, b,zy):
    #Á°ÆÂÆöÂõæÁâáÂêçÁß∞Âíå‰øùÂ≠òÂêçÁß∞
    regex = r"BioLab_(.*)_(.*)"
    mysearch = re.search(regex,file)
    if mysearch:
        plot_name = mysearch.group(1)
    path_plot = os.path.join(path,"plot",plot_name+"_"+zy)
    title_plot = plot_name+"_"+ zy
    path_plot_sum = os.path.join(path,"plot",plot_name + "_"+ zy+"_Sum")
    tile_plot_sum = plot_name + "_"+ zy + "_Sum"

    #ËØªÂèñÊñá‰ª∂
    df = pd.read_csv(file)
    df_filter = df[df["part"]==zy]
    p1 = df_filter["p1-c"].loc[a:b]
    p2 = df_filter["p2-c"].loc[a:b]
    p3 = df_filter["p3-c"].loc[a:b]
    p4 = df_filter["p4-c"].loc[a:b]
    p5 = df_filter["p5-c"].loc[a:b]
    p6 = df_filter["p6-c"].loc[a:b]
    p7 = df_filter["p7-c"].loc[a:b]
    p8 = df_filter["p8-c"].loc[a:b]
    p9 = df_filter["p9-c"].loc[a:b]
    p10 = df_filter["p10-c"].loc[a:b]
    p11 = df_filter["p11-c"].loc[a:b]
    p12 = df_filter["p12-c"].loc[a:b]

    fig3 = plt.figure(3)
    # plt.ticklabel_format(style="plain")
    plt.plot(p1,label="p1",color="lime")
    plt.plot(p2,label="p2")
    plt.plot(p3,label="p3")
    plt.plot(p4,label="p4")
    plt.plot(p5,label="p5")
    plt.plot(p6,label="p6")
    plt.plot(p7,label="p7")
    plt.plot(p8,label="p8")
    plt.plot(p9,label="p9")
    plt.plot(p10,label="p10")
    plt.plot(p11,label="p11",color="cyan")
    plt.plot(p12,label="p12")

    
    plt.legend(loc=2, bbox_to_anchor=(1.05,1.0),borderaxespad = 0.)
    #Â∞ÜÂõæ‰æãÊîæÁΩÆÂú®ÂõæË°®Â§ñ
    plt.title(title_plot)
    plt.grid(linestyle="-.")
    # plt.ticklabel_format(style="plain")
    # plt.tight_layout()
    plt.savefig(path_plot,dpi=600,bbox_inches='tight') #‰øùËØÅ‰øùÂ≠òÁöÑÂÜÖÂÆπÊòæÁ§∫Âõæ‰æã

    fig4 = plt.figure(4)
    p_sum = p1+p2+p3+p4+p5+p6+p7+p8+p9+p10+p11+p12
    plt.plot(p_sum);
    plt.title(tile_plot_sum)
    plt.grid(linestyle="-.")
    plt.savefig(path_plot_sum,dpi=600,bbox_inches='tight')
    # plt.tight_layout()
    plt.show() #ÂÖà‰øùÂ≠òÂÜçÊòæÁ§∫Ôºå‰∏çÁÑ∂‰øùÂ≠òÁöÑÂÜÖÂÆπÂ∞±ÊòØÁ©∫ÁôΩÁöÑÂõæË°®
    plt.close() #ÂèäÊó∂ÂÖ≥Èó≠ÂõæË°®ÔºåËäÇÁúÅÂÜÖÂ≠ò
```
